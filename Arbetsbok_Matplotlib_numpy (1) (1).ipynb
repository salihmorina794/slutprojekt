### <p style="color:black;">1.1 Installera och importera NumPy</p>

Du behöver bara installera Numpy en gång med pip install eller conda. Du behöver importera NumPy varje gång du ska använda NumPy, konventionen är att importera med prefixet __np__.

```python
import pandas as pd
import matplotlib.pyplot as plt
import PySimpleGUI as sg

def open_pie_chart(df):
    C_names = df[df['Name'].str.startswith('C')]
    other_names = df[~df['Name'].str.startswith('C')]

    name_counts = [len(C_names), len(other_names)]
    labels = ['Starts with C', 'Other']
    colors = ['#ff9999', '#66b3ff']

    plt.pie(name_counts, labels=labels, colors=colors, autopct='%1.1f%%')
    plt.title('C_names compared to all the other names')
    plt.show()


def open_c_names(df):
    C_names = df[df['Name'].str.startswith('C')]
    c_names_str = '\n'.join(C_names['Name'].tolist())
    width = max(len(line) for line in c_names_str.split('\n')) * 10

    layout = [
        [sg.Text('C_names:', size=(width, None))],
        [sg.Multiline(c_names_str, size=(width, 10))],
        [sg.Button('Close')]
    ]

    window = sg.Window('C_names', layout, no_titlebar=True, location=(0, 0))

    while True:
        event, _ = window.read()

        if event == sg.WINDOW_CLOSED or event == 'Close':
            break

    window.close()


layout = [
    [sg.Text('Username:'), sg.Input(key='-USERNAME-')],
    [sg.Text('Password:'), sg.Input(key='-PASSWORD-', password_char='*')],
    [sg.Button('Login')]
]

window = sg.Window('Login', layout)

while True:
    event, values = window.read()

    if event == sg.WINDOW_CLOSED:
        break

    if event == 'Login':
        username = values['-USERNAME-']
        password = values['-PASSWORD-']

        if username == 'salih' and password == 'öppna123':
            window.close()

            df = pd.read_csv(r'C:\Users\salih.morina\Downloads\pokemondata.csv - pokemondata.csv.csv')
            

            layout = [
                [sg.Button('Open Pie Diagram')],
                [sg.Button('Open C_names')],
            ]

            window = sg.Window('Data Options', layout)

            while True:
                event, values = window.read()
                
                if event == sg.WINDOW_CLOSED:
                    break

                if event == 'Open Pie Diagram':
                    open_pie_chart(df)

                if event == 'Open C_names':
                    open_c_names(df)

            break

        sg.popup('Invalid login credentials', title='Error')
```
